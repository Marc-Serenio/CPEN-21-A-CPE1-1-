##Boolean Operators
###Boolean represent one of two values: True or False

```python
print(10>9)
print(10<9)
print(10==9)
print(10!=9)
```

```python
print(bool(True))
print(bool(False))
print(bool(1))
print(bool(0))
print(bool({}))
print(bool(None))
```

```python
def myFunction():return True
print(myFunction())
```

```python
##Boolean answer of a function
def myFunction():return True

if myFunction():
  print("Yes!")
else:
  print("No!")
```

```python
a=6
b=7
print(a>b)
print(a==b)
print(6==7)
print(6!=6)
```

###Python Operators

```python
print(10+5)
print(10-5)
print(10*5)
print(10/5)
print(10//5)   ##floor division
print(10/3)    ##quotient
print(10//3)   ##floor division
print(10%3)    ##modulo
print(10**2)   ##concatenation
```

```python
a=60  ##0011 1100, 0111 1000, 1111 0000
b=13
print(a&b)
print(a|b)
print(a<<1)
print(a<<2)
print(a>>1) #0011 1100, 0001 1110
```

```python
a+=2 #same as a=a+2, a=60+2, a=62
print(a)
```

###Logical Operators

```python
a=True
b=False
print(a and b)
print(a or b)
```

```python
a=60
b=13
(a>b) and (a<b)
(a>b) or (a<b)
not(a>b)
```
